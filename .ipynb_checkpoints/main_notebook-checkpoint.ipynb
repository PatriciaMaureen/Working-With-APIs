```python
import json
import requests
import pandas as pd
```

```python
#1. Cat Facts

response = requests.get('https://cat-fact.herokuapp.com/facts/random?animal_type=cat&amount=50')
response
data = pd.DataFrame(response.json())
data.head()
```

```python
data[['text']]
```

```python
# You can see that people put a lot of junk in there, too. But there still are some cat facts in here. I guess before you want to use the data you have to clean it.
# It was just a fun start. I do not really see a project here.
```

```python
#2. Crypto Currency

response = requests.get('https://api.coincap.io/v2/assets')
data = pd.DataFrame(response.json())
data
```

```python
from pandas import json_normalize 

results = response.json()
flattened_data = json_normalize(results["data"])
flattened_data

```

```python
# There is the possibility to get historical data, too. For a project you could analyze the correlation between different crypto currencies and check for certain events to figure out what drives the prices.
```

```python
# 3. Open Food Facts
#https://openfoodfacts.github.io/openfoodfacts-server/api/tutorial-off-api/

response = requests.get('https://world.openfoodfacts.net/api/v2/product/3017624010701?fields=product_name,nutriscore_data')

results = response.json()
data = pd.DataFrame(results)
data
```

```python
results = response.json()
flattened_data = json_normalize(results["product"])
flattened_data
```

```python
# You can also get a list of foods for example Orange Juices with a special rating. Then you could use the product code to get the details of each product as seen above.
```

```python
response = requests.get('https://world.openfoodfacts.net/api/v2/search?categories_tags_en=Orange Juice&nutrition_grades_tags=c&fields=code,nutrition_grades,categories_tags_en')
data = pd.DataFrame(response.json())
data
```

```python
results = response.json()
flattened_data = json_normalize(results["products"])
flattened_data
```
